```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
study_hours=np.array([2,3,4,5,6]).reshape(-1,1)
exam_scores=np.array([65,75,80,85,90])
```

```python
model=LinearRegression()
```

```python
model.fit(study_hours,exam_scores)
```

```python
b0=model.intercept_
b1=model.coef_[0]
print(f"intercept (b0):{b0}")
print(f"Slope(b1):{b1}")


```

```python
new_study_hours=np.array([7]).reshape(-1,1)
predicted_scores=model.predict(new_study_hours)
```

```python
import matplotlib.pyplot as plt
plt.scatter(study_hours,exam_scores,label="Actual Data")
plt.plot(study_hours,model.predict(study_hours),color="red",label="Regression line")
plt.scatter(new_study_hours,predicted_scores,color="green",label="Predictions")
plt.xlabel("Study Hours")
plt.ylabel("Exam Scores")
plt.legend()
plt.title("Linear Regression Example")
plt.show()
```
